```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_dataset.csv")

df.head()
```

Count Plot

```python
sns.countplot(x="Gender", data=df)

plt.title("Number of Male and Female Students")

plt.show()
```

Histogram

```python
sns.histplot(df["Math"], bins=5)

plt.title("Distribution of Math Marks")

plt.show()
```

Box Plot

```python
sns.boxplot(y="Science", data=df)

plt.title("Science Marks")

plt.show()
```

Violin Plot

```python
sns.violinplot(x="Gender", y="English", data=df)

plt.title("English Marks by Gender")

plt.show()
```

Scatter Plot

```python
sns.scatterplot(x="Math", y="Science", data=df)

plt.title("Math vs Science")

plt.show()
```

Bar Plot

```python
sns.barplot(x="Gender", y="Attendance", data=df)

plt.title("Average Attendance")

plt.show()
```

Heatmap

```python
corr = df.select_dtypes(include="number").corr()

sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")

plt.show()
```

Pair Plot

```python
sns.pairplot(df[["Math","Science","English","Attendance"]])

plt.show()
```

Change Style

```python
sns.set_style("darkgrid")

sns.countplot(x="Gender", data=df)

plt.title("Students by Gender")

plt.show()
```
