```python
import re
```

The search() function checks whether a word exists in the text.

```python
text = "Welcome to Python"

result = re.search("Python", text)

print(result)
```

Regex can identify email addresses from text.

```python
text = "Contact: suman@gmail.com"

emails = re.findall(r"\S+@\S+", text)

print(emails)
```

Extract a 10-digit phone number.

```python
text = "Call me at 9876543210"

phones = re.findall(r"\d{10}", text)

print(phones)
```

Extract dates in DD/MM/YYYY format.

```python

text = "Date: 22/07/2026"

dates = re.findall(r"\d{2}/\d{2}/\d{4}", text)

print(dates)
```

The sub() function replaces text.

```python
text = "I like Java"

new_text = re.sub("Java", "Python", text)

print(new_text)
```

Remove symbols such as @, #, !, etc.

```python
text = "Hello@# Python!!"

clean = re.sub(r"[^A-Za-z0-9 ]", "", text)

print(clean)
```

Replace multiple spaces with a single space.

```python
text = "Python     is     easy"

clean = re.sub(r"\s+", " ", text)

print(clean)
```
