The info() function provides information about the dataset such as the number of rows, columns, data types, and missing values.

```python
print(df.info())
```

The describe() function shows statistical information such as count, mean, minimum, maximum, and standard deviation.

```python
print(df.describe())
```

Missing values are empty cells in a dataset. The isnull() function checks for missing values.

```python
print(df.isnull().sum())
```

```python
# remove rows with missing values.
df = df.dropna()
```

```python
print(df.isnull().sum())
```

The groupby() function groups data based on a column and performs operations such as calculating averages or totals.

```python
data = {
    "Department": ["IT", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 45000, 48000]
}

df = pd.DataFrame(data)

print(df.groupby("Department")["Salary"].mean())
```

The merge() function combines two DataFrames using a common column.

```python
student = pd.DataFrame({
    "ID": [1, 2],
    "Name": ["Suman", "Aman"]
})

marks = pd.DataFrame({
    "ID": [1, 2],
    "Marks": [85, 90]
})

result = pd.merge(student, marks, on="ID")

print(result)
```

A Pivot Table summarizes data by grouping and applying an operation such as sum or average.

```python
data = {
    "Department": ["IT", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 45000, 48000]
}

df = pd.DataFrame(data)

print(df.pivot_table(values="Salary", index="Department", aggfunc="mean"))
```
